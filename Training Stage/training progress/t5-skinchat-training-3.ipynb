# Import Library

```python
# For data
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

# For model, tokenizer
from simplet5 import SimpleT5
import transformers
from transformers import AutoTokenizer, AutoConfig, T5ForConditionalGeneration

# For neccessary
import os
from tqdm.auto import tqdm
import warnings
warnings.filterwarnings("ignore")
```

# Load dataset

```python
train_df = pd.read_csv("/kaggle/input/t5-skinchat-data-tokenizer-preparation/data_train.csv")
test_df = pd.read_csv("/kaggle/input/t5-skinchat-data-tokenizer-preparation/data_test.csv")
train_df.shape, test_df.shape
```

```python
train_df['source_text'] = train_df['source_text'].astype(str)
train_df['target_text'] = train_df['target_text'].astype(str)
test_df['source_text'] = test_df['source_text'].astype(str)
test_df['target_text'] = test_df['target_text'].astype(str)
```

# Load Tokenizer

```python
MODEL_NAME = "t5-base"
MAX_VOCAB = 20000
```

```python
new_tokenizer = AutoTokenizer.from_pretrained("/kaggle/input/t5-skinchat-training-2/outputs/simplet5-epoch-6-train-loss-0.9248-val-loss-1.3602/")
```

# Load Model

```python
model = T5ForConditionalGeneration.from_pretrained("/kaggle/input/t5-skinchat-training-2/outputs/simplet5-epoch-6-train-loss-0.9248-val-loss-1.3602/")
```

```python
# instantiate
simplet5 = SimpleT5()

simplet5.model = model
simplet5.tokenizer = new_tokenizer
```

```python
total_params = sum(p.numel() for p in simplet5.model.parameters())
print(f"Tổng số tham số của mô hình: {total_params}")
simplet5.model
```

# Training model

```python
os.environ["TOKENIZERS_PARALLELISM"] = "false"
```

```python
simplet5.train(train_df=train_df,
               eval_df=test_df,
               source_max_token_len=100,
               target_max_token_len=250,
               batch_size=16, 
               max_epochs=10, 
               use_gpu=True,
               early_stopping_patience_epochs = 4,
               precision=16
              )
```
